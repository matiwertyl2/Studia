```julia
using PyPlot
```

```julia
plotdx = 100
plotMax = 100000000

function sumElement(T, n)
    return T(1)/T(n)
end

function sum(T, arr)
    s = sumElement(T, arr[1])
    results = []
    for i in 2:length(arr)
        k = T(arr[i])
        s += sumElement(T, k)
        if  i % plotdx == 0
            push!(results, T(s))
        end
    end
    return results
end

function sumWithCorrection(T, arr)
    s = sumElement(T, arr[1])
    c = T(0)
    results=[]
    for i in 2:length(arr)
        k = T(arr[i])
        y = T(c + sumElement(T, k))
        t = T(s + y)
        c = (s - t) + y
        s = t
        if i % plotdx == 0
            push!(results, T(s))
        end
    end
    return results
end


function sumMistakes(Type, arr, sumAlgorithm)
    setprecision(BigFloat, 1024)
    good = sum(BigFloat, arr)
    checked = sumAlgorithm(Type, arr)
    mistakes=[]
    for i in 1:length(good)
        push!(mistakes, Float64(abs(good[i]-checked[i])/good[i]))
    end
    return mistakes
end

function getPlotX()
    res = [1:(plotMax/plotdx);] .* plotdx
    return res
end
show()
```

```julia
fig, ax = subplots()

x = getPlotX()
y1 = sumMistakes(Float32, [1:plotMax;], sum)
y2 = sumMistakes(Float32, [1:plotMax;], sumWithCorrection)
y3 = sumMistakes(Float32, reverse([1:plotMax;]), sum)

ax[:plot](x, y1, "-", color="red", linewidth=1, alpha=0.6, label="Zwyk≈Çe sumowanie")
ax[:plot](x, y2, "-", color="blue", linewidth=1, alpha=0.6, label="Sumowanie z poprawkami")
ax[:plot](x, y3, "-", color="green", linewidth=1, alpha=0.6, label="Sumowanie w odwrotnej kolejnosci")
ax[:legend](bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
title("Float32")

show()
```
