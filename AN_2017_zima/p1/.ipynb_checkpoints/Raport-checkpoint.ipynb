```julia
function sum(T, arr)
    s = T(0)
    for K in arr
        k = T(K)
        si = T(k)
        s += 1/k/k
    end
    return s
end

function sumWithCorrection(T, arr)
    s = T(1)/T(arr[1])/T(arr[1])
    c = T(0)
    for K in arr[2:end]
        k = T(K)
        y = T(c + T(1)/k/k)
        t = T(s + y)
        c = (s - t) + y
        s = t
    end
    return s
end

function countSum(manual, T, n)
    setprecision(BigFloat, 1024)
    sCorrect = sum(BigFloat, [1:n;])
    if manual
        setprecision(BigFloat, T)
        T = BigFloat
    end
    s1 = sum(T, [1:n;])
    s2 = sum(T, reverse([1:n;]))
    s3 = sumWithCorrection(T, [1:n;])
    s4 = sumWithCorrection(T, reverse([1:n;]))

    d1 = T(abs(s1 - sCorrect))
    d2 = T(abs(s2 - sCorrect))
    d3 = T(abs(s3 - sCorrect))
    d4 = T(abs(s4 - sCorrect))
    if manual

    else
        println(T)
    end
    println("NORMAL ", d1)
    println("REVERSE ", d2)
    println("WITH CORRECTION ", d3)
    println("WITH CORRECTION REVERSED ", d4)
end



countSum(false, Float64, 10000)
countSum(false, Float32, 10000)
countSum(true, 128, 10000)
```
